```python
from  sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

flatten = lambda l: [item for sublist in l for item in sublist]
scale = lambda data: abs(1-np.array(flatten(scaler.fit_transform(np.array(data).reshape(-1, 1)))))

scale_inv = lambda data: np.array(flatten(scaler.fit_transform(np.array(data).reshape(-1, 1))))


scaler = MinMaxScaler()

mae_avg = [0.081266667, 0.121866667, 0.1354, 0.1444]
mae_var = [0.004401396, 0.008997716, 0.004538773, 0.012635973]


x = scale(mae_avg)
y = scale(mae_var)


fig, ax = plt.subplots()
ax.scatter(x[0], y[0], color='r', s=150,  marker='^', alpha=.7, label='Árvores de Decisão')
ax.scatter(x[1], y[1], color='b', s=150, alpha=.4, label='Aprendizagem Profunda')
ax.scatter(x[2], y[2], color='g', s=150, marker='s', alpha=.7, label='Máquinas Vetor de Suporte')
ax.scatter(x[3], y[3], color='k', s=150, marker='*', alpha=.7, label='Regressão Linear')
fig.suptitle('Formulação 1', fontsize=18)
plt.xlabel('Performance', fontsize=16)
plt.ylabel('Estabilidade', fontsize=16)
ax.legend()
ax.grid(True)
plt.show()
```
