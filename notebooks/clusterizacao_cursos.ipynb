# Clusterização de Cursos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
#Alterando configurações padrão dos plots
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
df = pd.read_csv('../data/preprocessed/preprocessed_edx_2014_enriched.csv',low_memory=False)
```

```python
humanities = ['Humanities', 'Social Sciences', 'History']
exact = ['Computer Science', 'Engineering', 'Chemistry', 'Electronics', '']
biologics = ['Health & Safety', 'Environmental Studies', 'Biology & Life Sciences']


keys = df.groupby(['course_id', 'year']).groups.keys()

print(keys)
courses = []
for keys in keys:
    course_id = keys[0]
    year= keys[1]
    course_df = df[df['course_id']==course_id]
    course_df = course_df[course_df['year']==year]
    
    effort = course_df.groupby(['course_id'])['course_effort'].unique()[0][0]
    
    courses.append({
        'year': year,
        'course_id': course_id,
        'registrations': course_df.groupby(['course_id'])['userid_DI'].size()[0],
        'views': course_df.groupby(['course_id'])['viewed'].sum()[0],
        'exploration': course_df.groupby(['course_id'])['explored'].sum()[0],
        'certification': course_df.groupby(['course_id'])['certified'].sum()[0],
        'effort': effort if not np.isnan(effort) else -1,
        'size': course_df.groupby(['course_id'])['course_length'].unique()[0][0],
        'level': 0 if course_df.groupby(['course_id'])['course_level'].unique()[0][0] == 'Introductory' else 1,
        'is_humanities': 1 if course_df.groupby(['course_id'])['course_subject'].unique()[0][0] in humanities else 0,
        'is_exact': 1 if course_df.groupby(['course_id'])['course_subject'].unique()[0][0] in exact else 0,
        'is_biologics': 1 if course_df.groupby(['course_id'])['course_subject'].unique()[0][0] in biologics else 0,
        'inconsistents': course_df.groupby(['course_id'])['incomplete_flag'].sum()[0],
 #       'grade_mean': course_df.groupby(['course_id'])['grade'].mean()[0],
 #       'grade_std': course_df.groupby(['course_id'])['grade'].std()[0],
 #       'grade_missing': course_df.groupby(['course_id']).agg({'grade': lambda x: x.isnull().sum()})['grade'][0]

    })
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, MinMaxScaler

course_names_year = pd.DataFrame([courses['course_id'], courses['year']]).T

courses_cluster = courses.drop(columns=['course_id', 'year'])


std = StandardScaler()
courses_cluster = pd.DataFrame(std.fit_transform(courses_cluster))

cltr = KMeans(n_clusters=2, random_state=42, verbose=1, algorithm='elkan')
pred = cltr.fit_predict(courses_cluster)

def print_clusters(courses, pred):
    groups = {}

    for idx, label in enumerate(pred):
        cy = courses.iloc[idx]
        if not label in groups:
            groups[label] = []
        groups[label].append(cy['course_id'] + '_' + str(cy['year']))

    return groups

print_clusters(course_names_year, pred)
```
